```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(2026)
```

```python
population_size = 100000
ture_mean_income = 75000

population_income = np.random.lognormal(mean=np.log(ture_mean_income),sigma=0.5, size=population_size)

print(population_income)
print(np.mean(population_income))
```

```python
sample_size = 500
ramdom_sample = np.random.choice(population_income, size=sample_size)
print(np.mean(ramdom_sample))
```

```python
biased_sample = population_income[population_income >80000]
observed_biased_sample = np.random.choice(biased_sample, size=sample_size)
print(np.mean(biased_sample))
```

# **Algorithm R**

```python
import random
stream = range(1,1001)
k = 5
reservoir = []
```

```python
for i, item in enumerate(stream):
    if i < k:
      reservoir.append(item)
    else:
      j= random.randint(0,i)
      if j < k:
        reservoir[j] = item

print(reservoir)
```

## **Biased Population Data**

```python
population = np.concatenate([np.zeros(6000),np.ones(4000)])
```

```python
phone_prob = np.where(population == 0, 0.10, 0.90)
has_phone = np.random.rand(len(population)) < phone_prob
sample_frame = population[has_phone]
```

# **The Heckman Logic**

```python
weights = np.where(sample_frame == 0, 1/0.1 , 1/0.99)
```

```python
weighted_mean = np.average(sample_frame, weights= weights)
print (weighted_mean)
```
